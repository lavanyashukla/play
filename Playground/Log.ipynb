# Exploring Logarithms in Python

## Notation
**b<sup>y</sup>=x**

can be represented in logarithmic notation as:

**log<sub>b</sub>x = y**

For example **10<sup>2</sup>=100**

```python
10**2
```

translates to **log<sub>10</sub>100 = 2**

```python
math.log10(100)
```

## Logs are essentially the inverse operation to exponent, so they undo the exponent.

```python
x=100
print(np.exp(x))
print(np.log(np.exp(x)))
```

## Bases

```python
# Base 10
math.log10(100)
```

```python
# Base e, or ln which is the natural log
math.log(100)
```

```python
# Base 2
math.log2(16)
```

```python
# Base n (=20)
math.log(400, 20)
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

x = np.linspace(0.01, 1000, 100)
y_10 = np.log10(x)

plt.figure(figsize=(15,15))
plt.plot(x, y_10, marker='o', markevery=10)

plt.legend(['ln', 'log10', 'log2'], loc='lower right')
plt.xticks(range(math.floor(min(x)), math.ceil(max(x))+1, 100))
plt.axhline(0, color='black', linewidth='0.5')
plt.axvline(0, color='black', linewidth='0.5')
plt.title("Base 10")
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

x = np.linspace(0.01, 100, 100)
y_2 = np.log2(x)

plt.figure(figsize=(15,15))
plt.plot(x, y_2, marker='o', markevery=2)

plt.legend(['ln', 'log10', 'log2'], loc='lower right')
plt.xticks(range(math.floor(min(x)), math.ceil(max(x))+1, 2))
plt.axhline(0, color='black', linewidth='0.5')
plt.axvline(0, color='black', linewidth='0.5')
plt.title("Base 2")
```

## Propoerties
### The logarithm of 1 is zero
For all bases: y is 0 where x is 1

### When base = x, the logarithm is always 1
For all bases: y is 1 where x = b

### The logarithm diverges towards negative infinity as x approaches zero
For all bases: log(x) becomes negative for x < 1 for any base, and diverges towards negative infinity as x approaches zero

### The logarithm diverges towards infinity as as x approaches infinity
For all bases: as x approaches infinity, log(x) diverges towards positive infinity

```python
x = np.linspace(0.01, 20, 1000)
y_e = np.log(x)
y_10 = np.log10(x)
y_2 = np.log2(x)

plt.figure(figsize=(15,15))
plt.plot(x, y_e,marker='o', markevery=10)
plt.plot(x, y_10, marker='o', markevery=10)
plt.plot(x, y_2, marker='o', markevery=10)

plt.legend(['ln', 'log10', 'log2'], loc='lower right')
plt.xticks(range(math.floor(min(x)), math.ceil(max(x))+1, 1))
plt.yticks(range(1, 5, 1))
plt.axhline(0, color='black', linewidth='0.5')
plt.axvline(0, color='black', linewidth='0.5')
```

## Picking a Base

**Picking a log base**

- The base you pick is highly dependent on your data.
- So if your data triples at each step, use log base 3. If it goes up by 7.5 at each step, use log base 7.5.
- Pick the factor by which your data multiplies at each step as your base.
