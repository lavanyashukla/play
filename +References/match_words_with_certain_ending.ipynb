## Preliminaries

```python
# Load regex package
import re
```

## Create some text

```python
# Create a variable containing a text string
text = 'Capitalism, Communism, Neorealism, Liberalism'
```

## Apply regex

```python
# Find any word ending in 'ism'
re.findall(r'\b\w*ism\b', text)

# Specific:
# \b     - start of the word
# \w*    - a word of any length
# ism\b  - with 'ism'at the end
```
