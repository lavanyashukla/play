## Create Data

```python
data = [34, 54, 50, 20, 20]
```

## any()

```python
# Return true if any element is True
any(data)
```

## all()

```python
# Return true if all elements are True
all(data)
```

## max()

```python
# Return max value
max(data)
```

## min()

```python
# Return the min value
min(data)
```

## sum()

```python
# Return the total value
sum(data)
```
