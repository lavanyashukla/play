## Preliminaries

```python
# Load library
import numpy as np
```

## Create Matrices

```python
# Create matrix
matrix_a = np.array([[1, 1, 1],
                     [1, 1, 1],
                     [1, 1, 2]])

# Create matrix
matrix_b = np.array([[1, 3, 1],
                     [1, 3, 1],
                     [1, 3, 8]])
```

## Add Matrices

```python
# Add two matrices
np.add(matrix_a, matrix_b)
```

## Subtract Matrices

```python
# Subtract two matrices
np.subtract(matrix_a, matrix_b)
```
