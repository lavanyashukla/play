## Preliminaries

```python
# Load libraries
from sklearn.preprocessing import MultiLabelBinarizer
import numpy as np
```

## Create Data

```python
# Create NumPy array
y = [('Texas', 'Florida'), 
    ('California', 'Alabama'), 
    ('Texas', 'Florida'), 
    ('Delware', 'Florida'), 
    ('Texas', 'Alabama')]
```

## One-hot Encode Data

```python
# Create MultiLabelBinarizer object
one_hot = MultiLabelBinarizer()

# One-hot encode data
one_hot.fit_transform(y)
```

## View Column Headers

```python
# View classes
one_hot.classes_
```
