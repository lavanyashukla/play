## Preliminaries

```python
# Load libraries
from sklearn.preprocessing import Binarizer
import numpy as np
```

## Create Data

```python
# Create feature
age = np.array([[6], 
                [12], 
                [20], 
                [36], 
                [65]])
```

## Option 1: Binarize Feature

```python
# Create binarizer
binarizer = Binarizer(18)

# Transform feature
binarizer.fit_transform(age)
```

## Option 2: Break Up Feature Into Bins

```python
# Bin feature
np.digitize(age, bins=[20,30,64])
```
